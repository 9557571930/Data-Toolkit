## **Q-1. What is NumPy, and why is it widely used in Python?**
   ###  **What is NumPy?**  
NumPy (Numerical Python) is a powerful open-source library in Python used for numerical computing. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these data structures efficiently.

### **Why is NumPy Widely Used?**  
1. **Efficient Handling of Arrays & Matrices**  
   - NumPy provides the `ndarray` object, which is more memory-efficient and faster than Python‚Äôs built-in lists.
   - Supports multi-dimensional arrays (1D, 2D, 3D, and more).

2. **Performance Optimization**  
   - Operations on NumPy arrays are much faster because they are implemented in C and use vectorized operations instead of loops.
   - It leverages SIMD (Single Instruction, Multiple Data) and parallel computing.

3. **Broadcasting Feature**  
   - NumPy allows operations on arrays of different shapes without explicit looping, making code cleaner and more efficient.

4. **Rich Mathematical Functions**  
   - Provides built-in mathematical functions like linear algebra (`numpy.linalg`), Fourier transforms (`numpy.fft`), statistics (`numpy.mean`, `numpy.std`), and random number generation (`numpy.random`).

5. **Integration with Other Libraries**  
   - Many popular Python libraries like Pandas, SciPy, scikit-learn, TensorFlow, and PyTorch use NumPy as their core dependency.

6. **Indexing and Slicing Capabilities**  
   - NumPy provides powerful indexing, slicing, and reshaping functionalities that make data manipulation easier.

7. **Handling of Missing or NaN Values**  
   - NumPy provides functions to handle missing data (`numpy.nan`, `numpy.isnan`).

### **Example Usage of NumPy**
```python
import numpy as np

# Creating a NumPy array
arr = np.array([1, 2, 3, 4, 5])

# Performing vectorized operations

# Creating a 2D array (matrix)
```

##**Q-2 How does broadcasting work in NumPy?**
### **How Does Broadcasting Work in NumPy?**  

**Broadcasting** is a powerful feature in NumPy that allows element-wise operations on arrays of different shapes **without** creating unnecessary copies of data. This makes computations **more efficient and memory-friendly**.

---

### **1. Broadcasting Rules**
When performing operations between NumPy arrays, they are broadcasted to a **compatible shape** according to these rules:

1. **Rule 1: If the arrays have different dimensions, add 1s to the left of the smaller array's shape**  
   Example: `(3,)` ‚Üí `(1,3)` (converted to a 2D row vector)

2. **Rule 2: If dimensions don‚Äôt match, the dimension with size `1` is stretched to match the larger size.**  
   Example: `(1,3)` and `(3,3)` ‚Üí `(3,3)`

3. **Rule 3: If the dimensions are incompatible and cannot be stretched, NumPy throws an error.**  
   Example: `(3,4)` and `(2,4)` ‚Üí ‚ùå *Incompatible shapes, cannot broadcast*

---

### **2. Examples of Broadcasting**
#### **Example 1: Scalar and Array**
```python
import numpy as np

arr = np.array([1, 2, 3])  # Shape (3,)
result = arr + 5  # Broadcasting scalar (converted to shape (1,))
print(result)  # Output: [6 7 8]
```
üìå **How it works?**  
Scalar `5` is broadcasted to `[5, 5, 5]` (same shape as `arr`) and added element-wise.

---

#### **Example 2: 1D and 2D Arrays**
```python
A = np.array([[1, 2, 3], [4, 5, 6]])  # Shape (2,3)
B = np.array([10, 20, 30])  # Shape (3,)

result = A + B  # Broadcasting B to (2,3)
print(result)
```
üîπ **Output:**
```
[[11 22 33]
 [14 25 36]]
```
üìå **How it works?**  
- `B` has shape `(3,)`, NumPy reshapes it to `(1,3)`.  
- Then, it is **broadcasted** to match `(2,3)` and added element-wise.

---

#### **Example 3: Column Vector and Row Vector**
```python
A = np.array([[1], [2], [3]])  # Shape (3,1)
B = np.array([10, 20, 30])  # Shape (3,)

result = A + B  # Broadcasting A to (3,3)
print(result)
```
üîπ **Output:**
```
[[11 21 31]
 [12 22 32]
 [13 23 33]]
```
üìå **How it works?**  
- `A` has shape `(3,1)`, and `B` has `(3,)` ‚Üí reshaped to `(1,3)`.  
- They are **broadcasted** to `(3,3)` for element-wise addition.

---

### **3. When Broadcasting Fails**
```python
A = np.array([[1, 2], [3, 4]])  # Shape (2,2)
B = np.array([1, 2, 3])  # Shape (3,)

result = A + B  # ‚ùå Error: shapes (2,2) and (3,) are not compatible
```
üìå **Why?**  
- `(2,2)` and `(3,)` do not follow the broadcasting rules. **Dimensions must be either the same or 1.**

---

### **4. Practical Applications of Broadcasting**
‚úÖ **Adding a constant to a matrix:**  
```python
matrix = np.array([[1, 2], [3, 4]])
print(matrix + 10)  # Broadcasting scalar
```

‚úÖ **Normalizing data (subtracting mean):**  
```python
data = np.array([[10, 20, 30], [40, 50, 60]])
mean = np.mean(data, axis=0)  # Shape (3,)
normalized_data = data - mean  # Broadcasting
```

‚úÖ **Element-wise multiplication of different shapes:**  
```python
A = np.array([[1], [2], [3]])  # Shape (3,1)
B = np.array([10, 20, 30])  # Shape (3,)
result = A * B  # Broadcasting
```

---

### **Conclusion**
- **Broadcasting allows operations on mismatched shapes without explicit looping.**
- **It follows three rules** (add dimensions, match 1s, raise error if incompatible).
- **It's widely used in scientific computing, machine learning, and data processing.**

##**Q-4 A Explain the use of the groupby() method in Pandas.**
### **Understanding the `groupby()` Method in Pandas**  

The **`groupby()`** method in Pandas is used for **grouping data** based on one or more columns and then performing aggregate operations (such as sum, mean, count, etc.) on those groups. It is similar to SQL's `GROUP BY` clause.

---

## **1. Basic Syntax**
```python
df.groupby(by="column_name").aggregation_function()
```
- `by`: Column(s) used to group the data.  
- `aggregation_function()`: Function like `sum()`, `mean()`, `count()`, etc.

---

## **2. Example: Grouping by a Single Column**
```python
import pandas as pd

# Sample data
data = {
    "Department": ["HR", "IT", "HR", "IT", "Finance", "HR", "Finance"],
    "Employee": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace"],
    "Salary": [50000, 60000, 55000, 75000, 65000, 52000, 70000]
}

df = pd.DataFrame(data)

# Grouping by Department and calculating the average salary
grouped_df = df.groupby("Department")["Salary"].mean()
print(grouped_df)
```
üîπ **Output:**
```
Department
Finance    67500.0
HR         52333.3
IT         67500.0
Name: Salary, dtype: float64
```
‚úÖ **Explanation:** The employees are grouped by the `Department` column, and the average salary is calculated for each group.

---

## **3. Grouping by Multiple Columns**
```python
grouped_df = df.groupby(["Department", "Employee"])["Salary"].sum()
print(grouped_df)
```
üîπ This groups data by **both Department and Employee**, then sums the salaries (though in this case, it's just the same salary value per employee).

---

## **4. Using Multiple Aggregate Functions**
```python
grouped_df = df.groupby("Department")["Salary"].agg(["mean", "sum", "count"])
print(grouped_df)
```
üîπ **Output:**
```
             mean    sum  count
Department                    
Finance  67500.0  135000      2
HR       52333.3  157000      3
IT       67500.0  135000      2
```
‚úÖ **Explanation:**  
- `mean` ‚Üí Average salary per department  
- `sum` ‚Üí Total salary per department  
- `count` ‚Üí Number of employees in each department  

---

## **5. Filtering Groups using `filter()`**
You can filter out groups based on a condition:
```python
df_filtered = df.groupby("Department").filter(lambda x: x["Salary"].mean() > 60000)
print(df_filtered)
```
üîπ This keeps only the groups where the **average salary is greater than 60,000**.

---

## **6. Iterating Over Groups**
You can loop through groups and process them separately:
```python
for dept, group in df.groupby("Department"):
    print(f"\nDepartment: {dept}")
    print(group)
```

---

## **Conclusion**
- `groupby()` is used to **split** data into groups, **apply** functions, and then **combine** results.  
- It is essential for **data analysis, summarization, and reporting** in Pandas.  
- You can use multiple aggregation functions to analyze grouped data efficiently.

##**Q-6 What are the differences between NumPy arrays and Python lists?**
### **Differences Between NumPy Arrays and Python Lists**  

Both **NumPy arrays** and **Python lists** are used to store and manipulate data, but they have key differences in terms of **performance, functionality, and memory efficiency**.  

| Feature        | **NumPy Arrays** | **Python Lists** |
|---------------|-----------------|-----------------|
| **Speed** | Faster due to optimized C implementation. | Slower due to Python‚Äôs dynamic typing. |
| **Memory Efficiency** | Uses contiguous memory, making operations efficient. | Stores references to objects, consuming more memory. |
| **Data Type Consistency** | Homogeneous (all elements of the same type). | Heterogeneous (can contain different data types). |
| **Functionality** | Supports **vectorized operations**, broadcasting, and advanced mathematical functions. | Requires explicit loops for mathematical operations. |
| **Performance in Large Datasets** | Much better due to optimized C-based backend. | Slower as lists grow in size. |
| **Built-in Functions** | Provides specialized methods like `np.mean()`, `np.sum()`, `np.dot()`. | Lacks built-in mathematical functions; requires loops. |
| **Dimensionality** | Supports multi-dimensional arrays (e.g., 2D, 3D arrays). | Only supports 1D lists, requiring nested lists for multi-dimensional data. |

---

### **1. Speed Comparison**
```python
import numpy as np
import time

# NumPy Array
arr = np.arange(1_000_000)  
start_time = time.time()
arr = arr * 2  # Vectorized operation
print("NumPy Time:", time.time() - start_time)

# Python List
lst = list(range(1_000_000))
start_time = time.time()
lst = [x * 2 for x in lst]  # Using list comprehension
print("List Time:", time.time() - start_time)
```
üîπ **NumPy is significantly faster** because it uses efficient memory storage and vectorized operations.

---

### **2. Memory Efficiency Comparison**
```python
import sys

list1 = [1, 2, 3, 4, 5]
array1 = np.array([1, 2, 3, 4, 5])

print("List Size:", sys.getsizeof(list1), "bytes")  # Larger due to storing references
print("NumPy Array Size:", array1.nbytes, "bytes")  # More compact storage
```
üîπ **NumPy arrays consume less memory** because they store elements in a contiguous block of memory.

---

### **3. Mathematical Operations**
#### **NumPy (Efficient Vectorized Operations)**
```python
arr = np.array([1, 2, 3, 4, 5])
print(arr * 2)  # [2 4 6 8 10]
```
#### **Python List (Requires Explicit Loop)**
```python
lst = [1, 2, 3, 4, 5]
print([x * 2 for x in lst])  # [2, 4, 6, 8, 10]
```
üîπ **NumPy is more readable and optimized for such operations.**

---

### **4. Multi-Dimensional Support**
#### **NumPy Array**
```python
matrix = np.array([[1, 2], [3, 4]])
print(matrix.shape)  # (2,2)
```
#### **Python List**
```python
matrix = [[1, 2], [3, 4]]  # Nested lists needed
print(len(matrix), len(matrix[0]))  # (2,2)
```
üîπ **NumPy makes handling multi-dimensional data easier.**

---

### **Conclusion**
| **Feature**        | **NumPy Arrays** | **Python Lists** |
|--------------------|-----------------|-----------------|
| ‚úÖ **Best for Numerical & Large Data Operations** | ‚úÖ Yes | ‚ùå No |
| ‚úÖ **Memory Efficient** | ‚úÖ Yes | ‚ùå No |
| ‚úÖ **Faster Performance** | ‚úÖ Yes | ‚ùå No |
| ‚úÖ **Supports Multi-dimensional Data** | ‚úÖ Yes | ‚ùå No |

üîπ **Use NumPy when working with numerical data, large datasets, and performance-critical applications.**  
üîπ **Use Python lists when working with small, heterogeneous data types or when flexibility is needed.**

##**Q-8 What does the term ‚Äúvectorized operation‚Äù mean in NumPy?**
### **What is a "Vectorized Operation" in NumPy?**  

A **vectorized operation** in **NumPy** refers to performing **element-wise computations** on entire arrays **without using explicit loops**. These operations are optimized using **low-level C and Fortran implementations**, making them much **faster and more efficient** than traditional Python loops.  

---

### **Why Use Vectorized Operations?**
‚úÖ **Faster Execution** ‚Äì NumPy operations are implemented in **C**, avoiding the overhead of Python loops.  
‚úÖ **More Memory Efficient** ‚Äì Uses **contiguous memory** for faster data access.  
‚úÖ **More Readable & Concise** ‚Äì Eliminates explicit loops, making code cleaner.  
‚úÖ **Takes Advantage of CPU Optimizations** ‚Äì Uses **SIMD (Single Instruction, Multiple Data)** for parallel processing.  

---

### **Example: Vectorized vs. Non-Vectorized Operation**  

#### ‚ùå **Without Vectorization (Using Loops)**
```python
import numpy as np
import time

arr = np.arange(1_000_000)
result = np.zeros_like(arr)

start_time = time.time()
for i in range(len(arr)):  # Explicit loop
    result[i] = arr[i] * 2
print("Loop Time:", time.time() - start_time)
```

#### ‚úÖ **With Vectorization (NumPy)**
```python
start_time = time.time()
result = arr * 2  # Vectorized operation
print("NumPy Time:", time.time() - start_time)
```
üîπ **Vectorized NumPy operations are significantly faster** than loops!

---

### **Common Vectorized Operations in NumPy**
| **Operation** | **NumPy Vectorized Code** | **Equivalent Python Loop** |
|--------------|-------------------------|-------------------------|
| Addition | `result = arr1 + arr2` | `for i in range(len(arr1)): result[i] = arr1[i] + arr2[i]` |
| Multiplication | `result = arr * 2` | `for i in range(len(arr)): result[i] = arr[i] * 2` |
| Square Root | `result = np.sqrt(arr)` | `for i in range(len(arr)): result[i] = math.sqrt(arr[i])` |
| Dot Product | `result = np.dot(arr1, arr2)` | `sum(a * b for a, b in zip(arr1, arr2))` |

---

### **Conclusion**
‚úî **Vectorized operations are faster and more efficient** than loops.  
‚úî **They allow easy mathematical operations on large datasets.**  
‚úî **They take full advantage of NumPy‚Äôs underlying C optimizations.**

##**Q-13 Why is handling missing data important in Pandas?**
### **Why is Handling Missing Data Important in Pandas?**  

**Missing data** can negatively impact **data analysis, machine learning models, and decision-making**. If not handled properly, it can lead to **incorrect insights, biased models, or even errors** in computations.  

---

### **Key Reasons for Handling Missing Data**  

‚úÖ **1. Ensures Data Integrity & Accuracy**  
   - Missing values can **distort statistical calculations** (e.g., mean, median, correlations).  
   - Example: If **50% of salary data** is missing, the **average salary** may be misleading.  

‚úÖ **2. Prevents Errors in Computations**  
   - Many **Pandas functions & ML algorithms** cannot handle NaN values directly.  
   - Example: `df.mean()` may return NaN if too many missing values exist.  

‚úÖ **3. Improves Machine Learning Model Performance**  
   - ML models **can't process missing values** unless handled properly.  
   - Example: Decision Trees handle missing data better than Linear Regression.  

‚úÖ **4. Avoids Biased Analysis**  
   - Ignoring missing data may lead to **biased trends** in data-driven decisions.  
   - Example: Missing survey responses could skew public opinion analysis.  

---

### **How to Handle Missing Data in Pandas?**  

#### **1Ô∏è‚É£ Detect Missing Data**
```python
import pandas as pd

# Sample Data
data = {'Name': ['Alice', 'Bob', 'Charlie', None],
        'Age': [25, None, 30, 28],
        'Salary': [50000, 60000, None, 70000]}

df = pd.DataFrame(data)

# Check for missing values
print(df.isnull().sum())  # Count of NaNs per column
```
üîπ **Output:**  
```
Name      1
Age       1
Salary    1
dtype: int64
```

---

#### **2Ô∏è‚É£ Removing Missing Values (`dropna()`)**
```python
df_cleaned = df.dropna()  # Removes rows with NaNs
```
‚úî **Use when missing values are rare and dropping rows won't impact analysis.**  

---

#### **3Ô∏è‚É£ Filling Missing Values (`fillna()`)**
```python
df_filled = df.fillna({'Age': df['Age'].mean(), 'Salary': df['Salary'].median()})
```
‚úî **Useful for numerical data ‚Äì replaces NaNs with mean, median, or mode.**  

---

#### **4Ô∏è‚É£ Forward/Backward Filling (`ffill()`, `bfill()`)**
```python
df.ffill(inplace=True)  # Forward-fill: Uses the last known value
df.bfill(inplace=True)  # Backward-fill: Uses the next available value
```
‚úî **Best for time-series data.**  

---

### **Conclusion**  
‚úî Handling missing data **prevents errors, ensures accuracy, and improves ML models**.  
‚úî Pandas provides **flexible methods** like `dropna()`, `fillna()`, and `ffill()`.  
‚úî The right method depends on **data type, context, and analysis goals**.

---

## **1Ô∏è‚É£ `map()` ‚Äì Element-wise Function Application (Series Only)**
- Used **only on Pandas Series** (single column or row).  
- Works **element-wise**, applying a function to each value.  
- Supports **custom functions, built-in functions, and dictionaries**.  

### **Example: Using `map()` on a Series**
```python

# Sample Series

```
üîπ **Output:**
```

1     4
2     9
3    16
4    25
dtype: int64
```

‚úî `map()` is **faster and simpler** for element-wise transformations.  

---

## **2Ô∏è‚É£ `apply()` ‚Äì Apply Function on Rows/Columns (DataFrame & Series)**
- Works on **both DataFrames and Series**.  
- Can apply functions **element-wise, row-wise, or column-wise**.  
- More flexible than `map()`.  

### **Example 1: Using `apply()` on a Series**
```python
# Apply a function to double each element
doubled = s.apply(lambda x: x * 2)
print(doubled)
```
üîπ **Same output as `map()` but more versatile.**

### **Example 2: Using `apply()` on a DataFrame (Row-wise & Column-wise)**
```python
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# Apply function column-wise
col_sum = df.apply(lambda x: x.sum(), axis=0)  # axis=0 ‚Üí Columns
print(col_sum)

# Apply function row-wise
row_sum = df.apply(lambda x: x.sum(), axis=1)  # axis=1 ‚Üí Rows
print(row_sum)
```

‚úî `apply()` is **more powerful** as it can be used on **both DataFrames and Series**.

## **üÜö `apply()` vs. `map()` ‚Äì Key Differences**
| Feature      | `map()` (Series Only) | `apply()` (Series & DataFrame) |
|-------------|----------------|----------------|
| **Works on**  | **Series only** | **Series & DataFrame** |
| **Function Type** | Element-wise | Element-wise, row-wise, column-wise |
| **Use with DataFrame** | ‚ùå No | ‚úÖ Yes |

| **Supports Row & Column Operations** | ‚ùå No | ‚úÖ Yes |

## **üîπ When to Use Which?**

## **1Ô∏è‚É£ Vectorized Operations (Avoiding Loops)**

### **Example: Vectorized Computation vs. Loop**
```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

# Vectorized operation (FAST)
squared = arr ** 2

squared_loop = [x**2 for x in arr]

print(squared)
```
‚úî **Faster Execution** due to **low-level optimizations**.  

---

## **2Ô∏è‚É£ Broadcasting (Automatic Shape Expansion)**
- NumPy automatically **expands smaller arrays** to match the shape of larger ones **without explicit loops**.  

### **Example: Adding a 1D Array to a 2D Array**
```python
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
vec = np.array([10, 20, 30])

result = arr_2d + vec  # Automatically broadcasts vec across rows
print(result)
```
üîπ **Output:**
```
[[11 22 33]
 [14 25 36]]
```
‚úî **Saves memory** by avoiding extra copies of the data.  

---

## **3Ô∏è‚É£ Advanced Indexing (Boolean & Fancy Indexing)**
- Allows complex selections using **boolean conditions** and **arrays of indices**.  

### **Example 1: Boolean Indexing (Filter Elements)**
```python
arr = np.array([10, 15, 20, 25, 30])
filtered = arr[arr > 20]  # Select elements greater than 20
print(filtered)
```

##**Q-21 Why is NumPy‚Äôs array slicing faster than Python‚Äôs list slicing?**
### **Why is NumPy‚Äôs Array Slicing Faster Than Python‚Äôs List Slicing?** üöÄ  

NumPy's array slicing is significantly faster than Python's list slicing due to **efficient memory management, contiguous storage, and vectorized operations**. Let‚Äôs break it down:

---

## **1Ô∏è‚É£ NumPy Uses Contiguous Memory Storage (C-Style Arrays)**
- **NumPy arrays** are stored in **contiguous blocks of memory** (like C arrays).  
- **Python lists**, on the other hand, store elements **as pointers to objects**, leading to **slower access** times.  

### **Example: Contiguous Memory in NumPy**
```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print(arr.strides)  # Shows the byte step size for moving to the next element
```
‚úî **Benefit:** Since elements are stored in adjacent memory locations, accessing and slicing NumPy arrays is much faster!  

---

## **2Ô∏è‚É£ NumPy Slicing is a View, Not a Copy**
- **Python lists create a new copy** when sliced, leading to **higher memory usage** and **slower operations**.  
- **NumPy slicing returns a "view" (not a copy), meaning it doesn‚Äôt create new data‚Äîjust references existing memory.**  

### **Example: NumPy View vs. Python Copy**
```python
# NumPy slicing creates a view
arr = np.array([1, 2, 3, 4, 5])
sliced_arr = arr[1:4]  # No new memory allocated

# Python list slicing creates a copy
lst = [1, 2, 3, 4, 5]
sliced_lst = lst[1:4]  # New memory allocated
```
‚úî **Benefit:** NumPy avoids unnecessary memory allocation, making slicing **much faster!**  

---

## **3Ô∏è‚É£ NumPy Uses SIMD and Vectorized Operations**
- **NumPy utilizes vectorized operations** (SIMD: **Single Instruction, Multiple Data**) to process multiple elements **in parallel**.  
- **Python lists perform element-wise operations sequentially**, making them slower.  

### **Example: NumPy Slicing vs. List Slicing Performance**
```python
import numpy as np
import time

# Large dataset
size = 10**7  
arr = np.arange(size)
lst = list(range(size))

# NumPy slicing
start = time.time()
sliced_arr = arr[:size//2]  # FAST
end = time.time()
print("NumPy slicing time:", end - start)

# List slicing
start = time.time()
sliced_lst = lst[:size//2]  # SLOW
end = time.time()
print("Python list slicing time:", end - start)
```
‚úî **Result:** NumPy slicing is much **faster** due to **optimized memory access** and **vectorized operations**.  

---

## **4Ô∏è‚É£ NumPy Uses Optimized Low-Level C Code**
- NumPy is implemented in **C and Fortran**, making its array operations highly optimized.  
- Python lists, being dynamically typed, require additional processing, leading to **slower execution**.  

---

## **üîπ Summary: Why is NumPy Slicing Faster?**
| Feature               | **NumPy Array Slicing** | **Python List Slicing** |
|-----------------------|------------------------|--------------------------|
| **Memory Storage**    | Contiguous (C-style)   | Pointers (non-contiguous) |
| **Slicing Efficiency**| Returns a **view** (no copy) | Creates a **new copy** |
| **Performance**       | **Fast (vectorized, SIMD)** | **Slow (loop-based)** |
| **Low-Level Optimization** | **Uses C and Fortran** | Pure Python (slower) |
